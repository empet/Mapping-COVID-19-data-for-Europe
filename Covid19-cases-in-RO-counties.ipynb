## Covid19 cases in  RO-counties

```python
import plotly.graph_objects as go
from plotly.subplots import make_subplots
import pandas as pd
import numpy as np
import json
```

```python
with open('date_22_iunie_la_12_51.json', 'r') as fp:  #https://datelazi.ro/
    jdata = json.load( fp)
```

```python
juds = ['AB', 'AR', 'AG', 'BC', 'BH', 'BN', 'BT', 'BV', 'BR', 'BZ', 'CS', 'CL', 'CJ', 
        'CT', 'CV', 'DB', 'DJ', 'GL', 'GR', 'GJ', 'HR', 'HD', 'IL', 'IS', 'IF', 'MM', 
        'MH', 'MS', 'NT', 'OT', 'PH', 'SM', 'SJ', 'SB', 'SV', 'TR', 'TM', 'TL', 'VS', 'VL', 'VN', 'B']
```

```python
dates = list(jdata['historicalData'].keys())[::-1][16:]
jdaily = {}
jdaily['date'] = dates[1:]
for jud in juds:
    jdaily[jud] = [jdata['historicalData'][dates[1]]['countyInfectionsNumbers'][jud]-\
                   jdata['historicalData'][dates[0]]['countyInfectionsNumbers'][jud]]
```

```python
for k in range(2, len(dates)):
    before = jdata['historicalData'][dates[k-1]]['countyInfectionsNumbers']
    today = jdata['historicalData'][dates[k]]['countyInfectionsNumbers']
    for jud in juds:
        jdaily[jud].append(today[jud]-before[jud]) if today[jud]-before[jud]>=0 else jdaily[jud].append(0)
```

```python
df = pd.DataFrame(jdaily)
df.head()
```

```python
ncols = 3
fig = make_subplots(rows=14, cols=ncols, horizontal_spacing=0.025,
                    vertical_spacing=0.03, subplot_titles=juds)
```

```python
for k, jud  in  enumerate(juds):
    j = k % ncols
    i = (k-j) // ncols
    fig.add_trace(go.Scatter(x=df['date'], y=df[jud], 
                             name='', line_color='RoyalBlue', mode='lines'), i+1, j+1)
```

```python
fig.update_layout(title_text='22 Iunie',
                  title_x=0.5,
                  width=1000, height=1500, showlegend=False,
                  hovermode='x'
                 )
fig.update_xaxes(tickfont_size=10)
fig.update_yaxes(tickfont_size=10, range=[0, 65]);
fig.update_layout(yaxis42_range=[0,77],
                  yaxis5_range=[0, 94],
                  yaxis35_range=[0, 249],
                  yaxis29_range=[0,74]);
```

```python
fig.update_annotations(font_size=12);
fig.layout.annotations[41].update(text='BUC');
```
